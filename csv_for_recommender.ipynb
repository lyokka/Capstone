```python
# load data
import pandas as pd
import numpy as np

def user_music_rate(filename):
    df = pd.read_csv(filename, engine = 'python')
    
    # drop music_id is 0
    df = df[df['music_id'] != 0]
    
    df['music_id'] = df["music_id"].astype('str')
    df['user_id'] = df["user_id"].astype('str')

    rate = df.play_len/df.music_len
    # replace invalid value
    rate = rate.replace([np.inf, -np.inf], -1)
    df['rate'] = rate
    # generate feature abnormal_rate; a song is repeated more than 30 times, drop abnormal
    df = df[(df['rate']>= 0) & (df['rate'] <= 30)]
    
    # generate rating
    df['rating'] = np.nan
    df.loc[df['rate']<=0.2,'rating'] = 1
    df.loc[(df['rate']>0.2) & (df['rate']<=0.5) ,'rating'] = 2
    df.loc[(df['rate']>0.5) & (df['rate']<=0.7) ,'rating'] = 3
    df.loc[(df['rate']>0.7) & (df['rate']<=0.8) ,'rating'] = 4
    df.loc[(df['rate']>0.8), 'rating'] = 5
    
    # return df[['user_id', 'music_id', 'rate']]
    return df[['user_id', 'music_id', 'rating']]
```

```python
file = ['0301_c1.csv', '0302_c1.csv', '0303_c1.csv']
df_for_recommender = [user_music_rate(name) for name in file]
```

```python
total = df_for_recommender[0].append([df_for_recommender[1], df_for_recommender[2]])
total.to_csv('0301_3_rating.csv', index=False)
```
